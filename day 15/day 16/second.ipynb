# Types of Graphs
1. plot - line graph, between two parameters
2. bar  - bar graph of single parameter
3. hist - histogram of a bunch of data i.e frequency fo single parameter
4. scatter - dot graph between two params
5. boxplot - stastical overview in form of box for a single param mean,outliners,quantiles(25%,75%),IQR
6. stem - more like bar graph,Lines instead of bars
7. violinplot- violin shaped stastical overview graph spread of data(more like hist) single     parameter     (optional cat_data)

```python
x=np.linspace(0,5,10)
x1=np.arange(1,11,3)
```

```python
x1
```

```python
y=x**3
```

```python
plt.plot(x,y)
#plt.show()
plt.title('plot')
```

```python
plt.bar(x,y)
plt.title('bar')
```

```python
plt.hist(y)
plt.title('hist')
```

```python
plt.scatter(x,y)
plt.title('scatter')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show
```

```python
plt.boxplot(y)
plt.title('boxplot')
```

```python
plt.stem(x,y)
plt.title('stem')
```

```python
plt.violinplot(y)
plt.title('violinplot')
```

```python
y1 = x**4
```

```python
plt.plot(x , y, x, y1)
plt.legend(['y=x**3','y1=x**4'])
plt.show()
```

```python
plt.plot(x ,y)
plt.plot(x,y1)
plt.legend(['y=x**3','y1=x**4'])
plt.show()
```

```python
fig= plt.figure()
plt.plot(x ,y,lw=3,marker='o', markerfacecolor='yellow')
plt.plot(x,y1, marker ='v', markersize=10)
plt.legend(['y=x**3','y1=x**4'])
plt.grid()
plt.show()
```
